```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup       # parse HTML and XML
import urllib.request               # Open urls

url = "https://en.wikipedia.org/wiki/Portland%2C_Oregon"
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, "lxml") # (html.parser) used to parse cml and html
```

```python
print(soup.prettify())
```

```python
all_tables = soup.find_all("table")
all_tables
```

```python
table = soup.find('table', class_='infobox geography vcard')
print(table.prettify())
```

```python
headers = []

for row in table.findAll('tr'):
    header = row.find('th')
    if (header):
        headers.append(header.text)
      
for h in headers:
    print(h)
    print()
```

```python
headers = []

for row in table.findAll('tr'):
    header = row.find('th')
    if (header):
        header = header.text.replace('•', '').lstrip()
        headers.append(header)
      
for h in headers:
    print(h)
    print()
```

```python
headers = []
data = []

for row in table.findAll('tr'):
    header = row.find('th')
    if (header):
        td = row.find('td')
        # Only append row if there is data
        if (td):
            header = header.text.replace('•', '').lstrip()
            headers.append(header)
            data.append(td.text) 
        
for d in data:
    print(d)
    print()
```

```python
assert(len(headers) == len(data))
```

```python
df = pd.DataFrame(headers, columns=['Categories'])
df['Data'] = data
df
```

```python
headers = ['Country', 'State', 'Founded', 'Mayor', 'Elevation', 'Population', 'Demonym(s)']
data = [None] * len(headers)

for row in table.findAll('tr'):
    header = row.find('th')
    if header:
        header = header.text.replace('•', '').lstrip()
        if header in headers:
            td = row.find('td')
            if (td):
                data[headers.index(header)] = td.text
        
for (x, y) in zip(headers, data):
    print(x, y)
```

```python
headers = ['Country', 'State', 'Founded', 'Mayor', 'Elevation', 'Population', 'Demonym(s)']
data = [None] * len(headers)

for row in table.findAll('tr'):
    header = row.find('th')
    if header:
        header = header.text.replace('•', '').lstrip()        
        if header in headers:
            td = row.find('td')
            if (td):
                data[headers.index(header)] = td.text  
        elif 'Population' in header:
            pop = row.findNext('tr')
            td = pop.find('td')
            if (td):
                data[headers.index('Population')] = td.text
        
for (x, y) in zip(headers, data):
    print(x, y)
```

```python
headers.insert(0, 'City')

city_name = soup.title.string.split(',', 1)
data.insert(0, city_name[0])

for (x, y) in zip(headers, data):
    print(x, y)
```
